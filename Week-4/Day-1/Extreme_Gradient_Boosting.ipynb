```python


cols_to_drop = ['url_google_index',

                  'domain_google_index',

                  'qty_vowels_domain',

                  'server_client_domain',

                  'tld_present_params',

                  'time_response',

                  'domain_spf',

                  'qty_ip_resolved',

                  'qty_nameservers',

                  'qty_mx_servers',

                  'ttl_hostname',

                  'url_shortened']



df=df.drop(cols_to_drop, axis=1)
```

```python

rows, columns = df.shape

print("Number of rows:", rows)

print("Number of columns:", columns)
```

```python
original_features=list(df.columns)
```

```python
original_features
```

```python
dataset_array= np.array(df)
```

```python
print(dataset_array)
```

```python
features_indices=[]
attributes = ['url','domain','file','params']
```

```python
new_dataset ={}
```

```python
for index,name in enumerate(original_features):#index represents index ofcurrent and the name represents the name
  if 'qty'in name and name.split('_')[-1] in attributes:#'qty' in substring
    features_indices.append([index,name.split('_')[-1]])#last part is from the attributes
  else:
    new_dataset[name]=dataset_array[:,index]#it assigns that the columns that dont meet
```

```python
for index,attribute in features_indices:
  if attribute == 'domain':
    if f'qty_char{attribute}' not in new_dataset.keys():#qty_char () 'url','domain','file','params' in the new_database
      new_dataset[f'qty_char{attribute}']=np.zeros(rows)#iterate new keyin the new_dataset

    new_dataset[f'qty_char{attribute}']+=dataset_array[:,index]
```

```python
df1 =pd.DataFrame(new_dataset).astype(int)
```

```python
df1[df1<-1]=-1#if the condition is true i will chane
df1
```

```python
#summary statistics of dataset

print(df1.describe())
```

```python
setup(data = df1, target = "phishing")#target variable
```

```python
#comparing and select the best data model

best_model = compare_models()
```

```python
tuned_model = tune_model(best_model,n_iter =1,optimize='F1')
```

```python
predictions = predict_model(tuned_model,data= df1)
```
